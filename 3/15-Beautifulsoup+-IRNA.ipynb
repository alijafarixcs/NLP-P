## Start Crawling

```python
import csv
finallist = []

Imagename = 1
for tag in soup.html.body.find(id="box4").div.ul.find_all('li'):
    try:
        Tagfinal = ""
        if tag.name != None:
            Text = tag.div.find("h3").find("a").text.encode("utf-8", errors='ignore').decode('UTF-8')

            URL  = "https://www.irna.ir/"+ tag.find("div").find("h3").find('a')['href']

            linksite = requests.get(URL)
            Extract = BeautifulSoup(linksite.text, 'html.parser')
            Topic = Extract.html.body.find(id="item").find_all("div")[2].find('ol').find_all("li")[0].text
            SubTopic = Extract.html.body.find(id="item").find_all("div")[2].find('ol').find_all("li")[1].text

            ImageURL = tag.find("figure").find("a").find('img')['src']
            response = requests.get(ImageURL)
            link = r"C:\Users\mohtat.WAN\Desktop\متن کاوی کد8\data\images" + str(Imagename) + ".jpg"
            with open(link, 'wb') as f:
                f.write(response.content)   
            print(Imagename)
            Imagename +=1
            Tagss  = Extract.html.body.find(id="item").find("div",class_="item-header d-lg-none").div.find_all('li')
            for i in Tagss:
                Tagg= i.a.text  + " ~ "
                Tagfinal += Tagg 
    except:
        pass
    finallist.append([Text,URL,Topic,SubTopic,Tagfinal])
from pandas import DataFrame

df = DataFrame (finallist,columns=['Text','URL',"Topic","SubTopic","Tagfinal"])


df.to_csv(r"D:\Text9\3\exportFinal.csv", sep='|', encoding='utf-8')

df

```
