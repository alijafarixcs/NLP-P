```python
SEED = 448
random.seed(SEED)
Textlist = random.sample(Textlist,3000)
```

```python
def clean(text):
    normalizer = Normalizer()
    # Remove URLs, User Mentions and Hashtags and Retweet and Number
    text = normalizer.normalize(text)
    text = ''.join(ch for ch, _ in itertools.groupby(text))
    # Remove Punctuation
    #text = re.sub(r'[^\w\s]', '', text)
    return text.strip()
```

```python
LabeledText = []
for i in Textlist:
    text = i.split("~")[0]
    label = i.split("~")[1]
    LabeledText.append((label, clean(text)))
LabeledText[1:10]
```

```python
def create_tfidf_training_data(Tuple_Duc):
    y = [d[0] for d in Tuple_Duc]
    corpus = [d[1] for d in Tuple_Duc]

    vectorizer = TfidfVectorizer(min_df=10, max_df=.5, encoding='utf-8', ngram_range=(1, 1),stop_words=stwd)
    X = vectorizer.fit_transform(corpus)
    return X, y
```

```python
X, y = create_tfidf_training_data(LabeledText)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0, random_state=42)
```

## SVM

### SVM Tune

```python
classifiersvm = SVC()

grid_param_SVM = dict(kernel=['linear', 'rbf','poly'],
                      C=[ 0.1, 1,10,.001],
                      gamma=[0.1, 1,10,.001])

clf = GridSearchCV(estimator=classifiersvm,
                         param_grid=grid_param_SVM,
                         scoring='accuracy',
                         cv=2,
                         n_jobs=-1)
```

```python
clf.fit(X_train, y_train)
best_parameters = clf.best_params_
print(best_parameters)
best_result = clf.best_score_
print(best_result)
```

```python
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r" % (mean, std * 2, params))
```

### SVM Save Model

```python
X_data = []
y_data = []
for i in LabeledText:
    X_data.append(i[1])
    y_data.append(i[0])
    
X_train_svm, X_test_svm, y_train_svm, y_test_svm = train_test_split(X_data, y_data, test_size = 0.2, random_state = 42)
svm = SVC(gamma=0.1,C=1, kernel='linear')
vec = TfidfVectorizer(min_df=10, max_df=0.5, encoding='utf-8', ngram_range=(1, 1),stop_words=stwd)
1
vec_clf = Pipeline([('vectorizer', vec), ('pac', svm)])
vec_clf.fit(X_train_svm, y_train_svm)

y_pred_svm = vec_clf.predict(X_test_svm)
print(y_pred_svm)

report = classification_report(y_test_svm, y_pred_svm)
print(report)

x = cross_val_score(vec_clf, X_data, y_data,cv=5)
print("cross_val is = ",x)

joblib.dump(vec_clf,r'D:\Text9\data\Svm-Topic.pkl', compress=9)
```
