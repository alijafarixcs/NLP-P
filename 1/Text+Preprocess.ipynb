## Stemming

```python
stemmer = Stemmer()
stemmer.stem('کتاب ها')
```

```python
for word in wordtokenizelist:
    wordstem = stemmer.stem(word)
    if wordstem != word:
        print(wordstem,"<-",word)
```

## Lemmatization

```python
lemmatizer = Lemmatizer()
lemmatizer.lemmatize('کتاب ها')
```

```python
for word in wordtokenizelist:
    wordlem = lemmatizer.lemmatize(word)
    if wordlem != word:
        print(wordlem,"<-",word)
```

```python
Textnorm = normalizer.normalize(Text).replace("\u200c","")
Textnorm2 = ''.join(ch for ch, _ in itertools.groupby(Textnorm))
Textnorm2.strip()
wordtokenizelist = word_tokenize(Textnorm2)

wordlist = []
for word in wordtokenizelist:
    if word not in Stopword:
        wordlist.append(lemmatizer.lemmatize(word))

dicwordcount = {}
for w in wordlist:
    if w not in dicwordcount.keys():
        dicwordcount[w] = 1
    else:
        dicwordcount[w] +=1
        
sorted_Dict = sorted(dicwordcount.items(), key=operator.itemgetter(1),reverse=True)
sorted_Dict
```
